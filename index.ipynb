A common python programming pattern to counting objects, produce histograms, or update statistics is to make calls to a dictionary as you iterate through a list. For example, given a list of words, you can create a dictionary to store counts and then iterate through the list of words, checking how many times each word has appeared using your dictionary, and updating the dictionary count now that you've seen that word again. The `dictionary.get()` method is very useful in doing this.
```python
import re 
import string 

def word_clean(x):
    regex = re.compile('[@_!#$%^&*()<>?/\|}{~:.,]') 
    result_list = []
    for i in range(len(x)):
        if regex.search(x[i]) == None and x[i] != "'" and x[i] != "[" and x[i] != "s": 
            result_list.append(x[i])
    return result_list
```

```python
import re
words_list = [t.strip().lower() for t in re.findall(r'\b.*?\S.*?(?:\b|$)', macbeth)]
words_list = word_clean(words_list)
words_list
```

```python
#Check for most common word
from collections import Counter
common_list = Counter([word for word in words_list]).most_common(25)
# common_words = {}
# for character in words_list:
#     common_words[character] = common_words.get(character, 0) + 1 

# common_words.most_common(25)
```
